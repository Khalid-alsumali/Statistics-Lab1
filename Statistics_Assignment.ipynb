```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import datasets
import plotly.express as px
```

```python
boston = datasets.load_boston()
print(boston.DESCR)
```

## Q1: Save the dataset into pandas dataframe

```python
df_boston= pd.DataFrame(data=boston.data,columns=boston.feature_names)
df_boston.head()

                  
```

## Q2: Use different histograms to plot features that have right, left and zero skew.

```python
px.histogram(df_boston['CRIM'])
# skew to L

```

```python
px.histogram(df_boston['RM'])
# normal
```

```python
px.histogram(df_boston['AGE'])
# skew to R
```

## Q3: Prove your previous result using mean, median and mode.

```python
mean = df_boston['CRIM'].mean()
print('mean = ',mean)
median = df_boston['CRIM'].median()
print('median = ',median)
mode = df_boston['CRIM'].mode()
print('mode =',mode)
```

```python
mean = df_boston['RM'].mean()
print('mean = ',mean)
median = df_boston['RM'].median()
print('median = ',median)
mode = df_boston['RM'].mode()
print('mode =',mode)
```

```python
mean = df_boston['AGE'].mean()
print('mean = ',mean)
median = df_boston['AGE'].median()
print('median = ',median)
mode = df_boston['AGE'].mode()
print('mode =',mode)
```
